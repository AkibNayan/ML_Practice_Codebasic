```python
import pandas as pd
df = pd.read_csv('C:/Users/hp/OneDrive/Documents/Machine Learning Practice/Datasets/5_one_hot_encoding/homeprices.csv')
df.head()
```

```python
dummies = pd.get_dummies(df.town)
dummies
```

```python
merged = pd.concat([df, dummies], axis='columns')
merged
```

```python
final = merged.drop(['town', 'west windsor'], axis='columns')
final
```

```python
from sklearn import linear_model
model = linear_model.LinearRegression()
```

```python
X = final.drop('price', axis='columns')
X
```

```python
y = final['price']
y
```

```python
model.fit(X, y)
```

```python
model.predict([[2800, 1, 0]])
```

```python
model.predict([[3400, 0, 0]])
```

```python
model.score(X, y)
```
